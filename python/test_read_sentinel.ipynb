```python
import os
import filament
import unittest
import matplotlib.pyplot as plt
import netCDF4
```

```python
datadir = "/data/SST/Sentinel/S3A_OL_2_WFR____20180716T105701_20180716T110001_20180717T180114_0179_033_265_2340_MAR_O_NT_002.SEN3/" 
datafile = os.path.join(datadir, "chl_oc4me.nc")
coordfile = os.path.join(datadir, "geo_coordinates.nc")
os.path.exists(datafile)
```

## Chlorophyll concentration

```python
with netCDF4.Dataset(coordfile, "r") as nc:
    lon = nc.get_variables_by_attributes(standard_name="longitude")[0][:]
    lat = nc.get_variables_by_attributes(standard_name="latitude")[0][:]
with netCDF4.Dataset(datafile, "r") as nc:
    chloro = nc.variables["CHL_OC4ME"][:]
```

```python
print("Field shape: {} X {}".format(chloro.shape[0], chloro.shape[1]))
```

```python
plt.pcolormesh(lon, lat, chloro)
plt.show()
plt.close()
```
