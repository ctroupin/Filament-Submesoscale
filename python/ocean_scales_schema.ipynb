```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Ellipse
import numpy as np
from matplotlib import rcParams
```

```python
rcParams.update({'font.size': 16})
```

```python
scales = {"submesoscale\ncurrents": (4, 3, 2, 2),
          "climate\nchange": (7, 8, 2, 3),
          "coastal\nupwelling": (5.5, 4, 2, 2),
          "turbulence": (1, 1., 3., 2.),
          "eddies": (5.5, 5.5, 2., 2),
          "thermohaline\ncirculation": (7.5, 6.5, 1.5, 2)
         }
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.subplot(111)
plt.plot(10, 11, "ko", ms=0.1)
plt.xlim(0, 10)
plt.ylim(0, 11)

for key, value in scales.items():
    ax.add_patch(Rectangle((value[0], value[1]), value[2], value[3], 
                           alpha=.3, linestyle='--', linewidth=1., edgecolor="k"))
    ax.text(value[0] + 0.5 * value[2], 
            value[1] + 0.5 * value[3], key, ha="center", va="center", fontsize=16)
    
    
plt.xticks(np.arange(0, 10), 
           (' ', '1 mm', '1 cm', '10 m', '100 m', 
            '1 km', '10 km', '100 km', '1000 km', '10000 km'))
plt.yticks(np.arange(0, 11), 
           (' ', '1 s', '1 min', '1 hour', '1 day', 
            '1 week', '1 month', '1 year', '10 years', '100 years', '1000 years'))
plt.xlabel("Spacial scales", fontsize=20)
plt.ylabel("Temporal\nscales", fontsize=20, rotation=0, ha="right")
plt.grid(linestyle="--", dashes=(5, 10), color=".75", alpha=4)
fig.autofmt_xdate()
plt.savefig("../figures/oceanscales.png", dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```
