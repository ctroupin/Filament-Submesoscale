```julia
using CairoMakie, GeoMakie
using NCDatasets
include("./config.jl")
```

```julia
datadir = "/home/ctroupin/data/SST/Yemen/"
datafile = joinpath(datadir, "SNPP_VIIRS.20241115T231200.L2.SST3.nc")
domain = [38, 50., 10., 16]
```

```julia
ds = NCDataset(datafile, "r")
lon = ds.group["navigation_data"]["longitude"][:,:]
lat = ds.group["navigation_data"]["latitude"][:,:]
SST = ds.group["geophysical_data"]["sst_triple"][:,:]
date = ds.attrib["time_coverage_start"]
figtitle = ds.attrib["title"]
print(ds)
close(ds)
```

```julia
fig = Figure(size=(1000, 750))

ga = GeoAxis(
    fig[1, 1],
    title = "$(figtitle)",
    dest = "+proj=merc"
)

sf = surface!(ga, lon, lat, SST, shading = NoShading, interpolate = false, colormap = Reverse(:RdYlBu), colorrange = (25., 30.))
Colorbar(fig[2, 1], sf, vertical = false, label = "(°C)")
# add_coast!(ga, coordscoast)
resize_to_layout!(fig)
xlims!(ga, domain[1], domain[2])
ylims!(ga, domain[3], domain[4])
save("../figures/SST_yemen.png", fig)
```
